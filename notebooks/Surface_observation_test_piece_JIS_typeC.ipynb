```python
import cv2
import matplotlib.pyplot as plt
```

```python
im = cv2.imread('./image/1.jpg')
plt.imshow(im)
plt.axis('off')
```

```python
img_src = cv2.imread("./image/1.jpg", 1)
img_gray = cv2.cvtColor(img_src, cv2.COLOR_BGR2GRAY)
thresh = 100 
max_pixel = 255
ret, img_dst = cv2.threshold(img_gray,
                                 thresh,
                                 max_pixel,
                                 cv2.THRESH_BINARY)
plt.imshow(img_dst)
plt.set_cmap('gray')
plt.axis('off')
```

```python
im = cv2.imread('./image/3.jpg')
plt.imshow(im)
plt.axis('off')
```

```python
thresh = 100 
```

```python
import numpy as np
# kernel = np.ones((5,5),np.float32)/25
kernel = np.ones((5,5),np.float32)/20
dst = cv2.filter2D(im,-1,kernel)

plt.imshow(im)
plt.axis('off')
plt.show()
ret, img_dst = cv2.threshold(im,
                                 thresh,
                                 max_pixel,
                                 cv2.THRESH_BINARY)
plt.imshow(img_dst)
plt.set_cmap('gray')
plt.axis('off')
plt.show()

plt.imshow(dst)
plt.axis('off')
plt.show()
ret, img_dst = cv2.threshold(dst,
                                 thresh,
                                 max_pixel,
                                 cv2.THRESH_BINARY)
plt.imshow(img_dst)
plt.set_cmap('gray')
plt.axis('off')
```
