```python
import numpy as np
import pandas as pd
df = pd.read_csv("out.csv",header=None)
```

```python
df 
```

```python
X = df.values[:,0]
Y = df.values[:,1]
Z = df.values[:,2]
```

```python
import matplotlib.pyplot as plt
plt.scatter(X,Y,s=50,c=Z)
```

```python
dX = (np.diff(X))
dY = (np.diff(Y))
travel = pd.DataFrame({'r':np.sqrt(np.square(dX)+np.square(dY)),'Z':Z[1:]})
```

```python
Z = np.array(travel.groupby('Z').sum().r.index)
Z
```

```python
R = travel.groupby('Z').sum().values
R
```

```python
plt.title('Travel distance at each layer')
plt.plot(Z,R)
plt.xlabel('Z[mm]')
plt.ylabel('travel distance[mm]')
```

```python
plt.hist(travel.groupby('Z').sum().values)
plt.title('Probability Density ')
```

```python
travel.groupby('Z').sum().plot(kind='density')
plt.title('Probability density function')
plt.xlabel('travel distance')
plt.ylabel('Probability density')
```
